```python
x = np.linspace(0,100,50)
y = 2.5 * x + np.random.normal(size =x.size)
```

```python
x
```

```python
y
```

```python
A = np.vstack([x, np.ones(x.size)]).T
A
```

```python
m, c = np.linalg.lstsq(A,y,rcond=None)[0]
```

```python
m
```

```python
c
```

```python
pendiente = m*x + c
```

```python
pendiente
```

```python
plt.plot(x, pendiente,'r',label='Regresion Lineal')
plt.scatter(x,y,label='Datos')
plt.legend()
plt.show()
```

```python
url_csv = "https://raw.githubusercontent.com/JJTorresDS/stocks-ds-edu/main/stocks.csv"
```

```python
stock_df = pd.read_csv(url_csv)
stock_df.head()
```

```python
nombres_mapeador = {
    "MCD": "McDonalds",
    'SBUX': 'Starbucks',
    'GOOG': 'Google',
    'AMZN': 'Amazon',
    'MSFT': 'Microsoft',
    'JPM': 'JPMorgan',
    'BAC': 'BankOfAmerica',
    'C': 'Citigroup',
    'MAR': 'Marriott',
    'HLT': 'Hilton',
    'RCL': 'RoyalCaribbean',
    'V': 'Visa',
    'MA': 'Mastercard',
    'PYPL': 'PayPal'
}
```

```python
stocks = stock_df.rename(columns=nombres_mapeador)
stocks.info()
```

```python
stocks['dates'] = pd.to_datetime(stocks.formatted_date) # stocks["formatted_date"]

stocks.info()
```

```python
stocks = stocks.set_index('dates')
stocks.drop(columns="formatted_date", inplace=True)
stocks.head()
```

```python
stocks.info()
```

```python
stocks.loc['2016-03-01':'2016-04-30']
```

```python
stocks.iloc[1:10].describe().T
```

```python
plt.figure(figsize=(16,9))

for column in stocks.columns:
  plt.plot(stocks.index, stocks[column], label=column)

# elementos de la graficaq AKA --> amor
plt.title('Cotizaciones de Acciones')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

```python
stocks.loc['2021-11-01']
```

```python
rendimiento = (stocks.iloc[-1,1:] - stocks.iloc[0,1:]) / stocks.iloc[0,1:]
```

```python
rendimiento.sort_values(ascending=False)
```

```python
returns = stocks.pct_change().dropna()
returns.head()
```

```python
plt.figure(figsize=(16, 9))
for column in returns.columns:
    plt.plot(returns.index, returns[column], label=column)

plt.title('Rendimientos Mensuales por Compañía')
plt.xlabel('Fecha')
plt.ylabel('Rendimiento (%)')
plt.legend(loc='upper right', fontsize='small', ncol=3)
plt.grid(True)
plt.tight_layout()
plt.show()

```
