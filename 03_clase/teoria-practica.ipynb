```python
x = np.linspace(0,10,50)
y = 2.5 * x + np.random.normal(size = x.size)
```

```python
# para hacer una regresion lineal --> que matematicamente se obtiene por medio de la tecnica de cuadrados minimos

# Supn que tienes tus datos en x e y
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

pendiente=m*x + c
```

```python
plt.plot(x,pendiente,'r', label = 'Regresion lineal')
plt.scatter(x,y,label="Datos")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresin Lineal con NumPy')
plt.legend()
plt.grid(True)
plt.show()
```

# Hands on lab

```python
url_csv = "https://raw.githubusercontent.com/JJTorresDS/stocks-ds-edu/main/stocks.csv"
```

```python
# lectura de archivo 
import pandas as pd
```

```python
stocks = pd.read_csv(url_csv, sep=",")
stocks.head()
```

```python
nombres_map = {
    'MCD': 'McDonalds',
    'SBUX': 'Starbucks',
    'GOOG': 'Google',
    'AMZN': 'Amazon',
    'MSFT': 'Microsoft',
    'JPM': 'JPMorgan',
    'BAC': 'BankOfAmerica',
    'C': 'Citigroup',
    'MAR': 'Marriott',
    'HLT': 'Hilton',
    'RCL': 'RoyalCaribbean',
    'V': 'Visa',
    'MA': 'Mastercard',
    'PYPL': 'PayPal'
}

# Renombrar columnas
stocks = stocks.rename(columns=nombres_map)
```

```python
stocks
```

```python
stocks['dates'] = pd.to_datetime(stocks['formatted_date'])
stocks = stocks.set_index('dates')
stocks.drop(columns=["formatted_date"],inplace=True)
```

```python
stocks.head()
```

```python
stocks.tail()
```

## Stocks grafica

```python
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'yellow', 'black', 'teal']
```

```python
# size de la imagen
plt.figure(figsize=(16, 9))
plt.gca().set_prop_cycle(color=colors)

for column in stocks.columns:
    plt.plot(
        stocks.index,
        stocks[column],
        label=column
    )
# Graficas elemetnos
plt.title('Cotizaciones de Acciones')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
stocks.iloc[-1, 1:]
```

```python
stocks.iloc[0, 1:]
```

```python
# Calcular rendimiento total
rendimientos = (stocks.iloc[-1, 1:] - stocks.iloc[0, 1:]) / stocks.iloc[0, 1:]
```

```python
acciones_recomendadas = rendimientos.sort_values(ascending=False).head(2)
print(acciones_recomendadas)
```

```python
returns = stocks.pct_change().dropna()
```

```python
plt.figure(figsize=(16, 9))
for column in returns.columns:
    plt.plot(returns.index, returns[column], label=column)

plt.title('Rendimientos Mensuales por Compaa')
plt.xlabel('Fecha')
plt.ylabel('Rendimiento (%)')
plt.legend(loc='upper right', fontsize='small', ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
returns['Amazon'].plot(kind='bar', figsize=(10,6), color='skyblue')

plt.title('Rendimientos Mensuales de Amazon')
plt.xlabel('Fecha')
plt.ylabel('Rendimiento')
plt.grid(True)
plt.show()
```

```python
# Nmero de acciones
n_stocks = len(returns.columns)

# Definir tamao de la grilla (filas y columnas) para los subplots
# Por ejemplo, 4 columnas y tantas filas como necesites
cols = 4
rows = (n_stocks + cols - 1) // cols  # redondeo hacia arriba

fig, axes = plt.subplots(rows, cols, figsize=(20, 4 * rows), sharex=True, sharey=True)
axes = axes.flatten()  # Para iterar fcilmente aunque sea 2D

# Graficar cada stock en un subplot individual
for i, stock in enumerate(returns.columns):
    axes[i].bar(returns.index, returns[stock], color='skyblue')
    axes[i].set_title(stock)
    axes[i].grid(True)
    # Opcional: no mostrar etiquetas x para todos menos el ltimo fila
    if i // cols < rows - 1:
        axes[i].set_xticklabels([])
    else:
        axes[i].set_xticklabels(returns.index.strftime('%Y-%m'), rotation=45, fontsize=8)

# Ocultar subplots vacos si los hay
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle('Rendimientos Mensuales por Compaa', fontsize=16)
fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
